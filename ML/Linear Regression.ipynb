Define cost function:
$$C = \frac{1}{2m} \sum_{i=1}^m (y_{obs}^i - y_{pred}^i)^2$$

```python
def CostFunction(obs,mod):
    cost = np.sum( (obs - mod)**2 ) / (2.*len(obs))
    return cost
```

The following function creates function to plot cost across a range of values

```python
def plotCost():
    PlotVec = []
    for theta_0 in np.linspace(-1000,2000,50):
        for theta_1 in np.linspace(-0.5,0.5,50):
            y_predicted_test = LinFunc(df.X,theta_0,theta_1)
            cost = CostFunction(df.Y,y_predicted_test)
            PlotVec.append([theta_0,theta_1,cost])
    PlotVec = np.array(PlotVec)
    plt.tricontour(PlotVec[:,0],PlotVec[:,1],PlotVec[:,2],80)
    pass
```

## Gradient Descent

Define two gradient functions:
$$g_0 = \frac{1}{m} \sum_{i=1}^m (y_{pred}^i - y_{obs}^i)$$
$$g_1 = \frac{1}{m} \sum_{i=1}^m (y_{pred}^i - y_{obs}^i).x^i$$

```python
def g0(obs,mod):
    g0= np.sum( mod - obs ) / float(len(obs))
    return g0

def g1(obs,mod,x):
    g1 = np.sum( (mod - obs) * x ) / float(len(obs))
    return g1
```

Finish gradient descent algorithm where:
$$\theta_0 := \theta_0 - \alpha_0 g_0$$
$$\theta_1 := \theta_1 - \alpha_1 g_1$$

```python
def GradDescent(theta_0,theta_1,alpha_0,alpha_1):
    converged = False
    
    #theta_history records values for plotting
    theta_history = []               
    theta_history.append([theta_0,theta_1])
    #Set last_cost to check for convergence
    last_cost = float("inf")
    
    while not converged:
        #Calculate prediction for parameter values
        df.Ypredicted = LinFunc(df.X,theta_0,theta_1)
        
        #Update parameters
        theta_0 = theta_0 - alpha_0 * g0(df.Y,df.Ypredicted)
        theta_1 = theta_1 - alpha_1 * g1(df.Y,df.Ypredicted,df.X)
    
        #Checking for convergence
        new_cost = CostFunction(df.Y,df.Ypredicted)
        if (new_cost-last_cost)**2 < 10-3 or np.shape(theta_history)[0] > 2000:
            converged = True
            
        last_cost = new_cost    
        theta_history.append([theta_0,theta_1])
        
    print "Minimum found at ", theta_history[-1], " in ", np.shape(theta_history)[0], " steps"
    return np.array(theta_history)
```

Set initial guess for parameters and look at alpha values, then run Gradient Descent algorithm

```python
theta_0_guess = 500. 
theta_1_guess = -0.15 
alpha_0 = 1.
alpha_1 = 1e-7

theta_history = GradDescent(theta_0_guess,theta_1_guess,alpha_0,alpha_1)
```

Use following to plot cost surface and parameter update history

```python
plotCost()
plt.plot(theta_history[:,0],theta_history[:,1])
plt.xlim([-1000,2000])
plt.ylim([-0.5,0.5])
```

Plot scatter and predicted function as before

```python
df.Ypredicted = LinFunc(df.X, theta_history[-1][0], theta_history[-1][1])
plt.scatter(df.X,df.Y)
plt.plot(df.X,df.Ypredicted)
```
